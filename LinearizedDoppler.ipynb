## Linearized Doppler operator

Here we'll show how to Taylor expand the Doppler operator so that we can model the velocity-dependent stellar spectrum using spherical harmonics. Specifically, we wish to compute the derivatives 
$\frac{\mathrm{d}\,I}{\mathrm{d}\,\beta}$,
$\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\beta^2}$,
...,
where $I = I(\lambda_0)$ is the spectral specific intensity at the rest frame wavelength $\lambda_0$ and $\beta = \frac{v}{c}$ is the relativistic beta. We assume that we know how to compute the derivatives of $I$ with respect to the rest frame wavelength $\lambda_0$, so we'll use the chain rule to figure out the derivatives with respect to $\beta$.

The non-relativistic Doppler shift is given by

$
\frac{\lambda - \lambda_0}{\lambda_0} = \beta
$

where $\lambda_0$ is the rest frame wavelength. Solving for $\lambda_0$, we have

$
\lambda_0 = \frac{\lambda}{1 + \beta} \equiv D^{-1}(\lambda, \beta)
$

where $D^{-1}(\lambda, \beta)$ is the inverse Doppler operator. Given a wavelength in the *shifted* frame, the inverse Doppler operator returns the rest frame wavelength.

### First derivative

To compute the derivative of $I$ with respect to $\beta$, we simply write 

$
I = I(\lambda_0) = I(D^{-1}(\lambda, \beta))
$

and differentiate using the chain rule:

$
\frac{\mathrm{d}\,I}{\mathrm{d}\,\beta} = 
    \frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}
    \frac{\mathrm{d}\,\lambda_0}{\mathrm{d}\,D^{-1}}
    \frac{\mathrm{d}\,D^{-1}}{\mathrm{d}\,\beta}
$

The first derivative on the RHS is assumed to be known. The second one is trivially equal to one, and the third one is

$
\frac{\mathrm{d}\,D^{-1}}{\mathrm{d}\,\beta} = -\frac{\lambda}{(1 + \beta)^2} \\
\,\,\,\,\,\,\,\,\,\,\,\, = -\frac{\lambda_0}{(1 + \beta)}
$

so the derivative with respect to $\beta$ is

$
\frac{\mathrm{d}\,I}{\mathrm{d}\,\beta} = -\frac{\lambda_0}{(1 + \beta)} \frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}
$

We plan on using these derivatives to Taylor expand the Doppler operator about $\beta = 0$, so that is where we evaluate them. The desired derivative (evaluated at $\beta = 0$) is thus

$
\frac{\mathrm{d}\,I}{\mathrm{d}\,\beta}\bigg|_{\beta=0} = -\lambda_0 \frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}
$

### Second derivative

Let's differentiate the expression for the first derivative. Setting $f(\lambda_0) = \frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}$, we have

$
\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\beta^2} = -\frac{\mathrm{d}\,}{\mathrm{d}\,\beta}\left[\frac{\lambda_0}{(1 + \beta)} \, f(\lambda_0)\right] \\
\,\,\,\,\,\,\,\,\, = \frac{\lambda_0}{(1+\beta)^2}f(\lambda_0) - \frac{\lambda_0}{(1 + \beta)} \frac{\mathrm{d}\,f}{\mathrm{d}\,\beta}
$

By comparison with the equations used to compute the first derivative, the derivative of $f$ is simply

$
\frac{\mathrm{d}\,f}{\mathrm{d}\,\beta} = -\frac{\lambda_0}{(1 + \beta)} \frac{\mathrm{d}\,f}{\mathrm{d}\,\lambda_0} \\
\,\,\,\,\,\,\, = -\frac{\lambda_0}{(1 + \beta)} \frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2}
$

so

$
\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\beta^2} = 
\frac{\lambda_0}{(1+\beta)^2}\frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}
+ \frac{\lambda_0^2}{(1 + \beta)^2} \frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2}
$

Evaluating at $\beta=0$, we have

$
\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\beta^2}\bigg|_{\beta=0} =
\lambda_0\frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0} + \lambda_0^2 \frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2}
$

### Third derivative
Following the same operations as above, we have

$
\frac{\mathrm{d^3}\,I}{\mathrm{d}\,\beta^3} = 
-\frac{\lambda_0}{(1+\beta)^3}\left[
    2\frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0} + 
    2\lambda_0\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2} + 
    \lambda_0^2 \frac{\mathrm{d^3}\,I}{\mathrm{d}\,\lambda_0^3}
\right]
$

and

$
\frac{\mathrm{d^3}\,I}{\mathrm{d}\,\beta^3}\bigg|_{\beta=0} =
-2\lambda_0 \frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0}
-2\lambda_0^2 \frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2}
-\lambda_0^3 \frac{\mathrm{d^3}\,I}{\mathrm{d}\,\lambda_0^3}
$

### Fourth derivative
Following the same operations as above, we have

$
\frac{\mathrm{d^3}\,I}{\mathrm{d}\,\beta^3} = 
\frac{\lambda_0}{(1+\beta)^4}\left[
    6\frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0} + 
    8\lambda_0\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2} + 
    5\lambda_0^2 \frac{\mathrm{d^3}\,I}{\mathrm{d}\,\lambda_0^3} +
    \lambda_0^3 \frac{\mathrm{d^4}\,I}{\mathrm{d}\,\lambda_0^4}
\right]
$

and

$
\frac{\mathrm{d^3}\,I}{\mathrm{d}\,\beta^3}\bigg|_{\beta=0} =
    6\lambda_0\frac{\mathrm{d}\,I}{\mathrm{d}\,\lambda_0} + 
    8\lambda_0^2\frac{\mathrm{d^2}\,I}{\mathrm{d}\,\lambda_0^2} + 
    5\lambda_0^3 \frac{\mathrm{d^3}\,I}{\mathrm{d}\,\lambda_0^3} +
    \lambda_0^4 \frac{\mathrm{d^4}\,I}{\mathrm{d}\,\lambda_0^4}
$

### Let's validate these expressions numerically

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

First, let's create a synthetic Gaussian absorption line and analytically compute its derivatives with respect to wavelength:

```python
# A really wide Gaussian absorption line
nwav = 3000
lam = np.linspace(450, 600, nwav)
amp = 0.1
mu = 525
sig = 20
I = 1 - amp * np.exp(-(lam - mu) ** 2 / sig ** 2)

# Its derivatives with respect to wavelength
dIdl = np.empty((4, nwav))
dIdl[0] = 2 / sig ** 2 * (1 - I) * (lam - mu)
dIdl[1] = 2 / sig ** 2 * (1 - I - dIdl[0] * (lam - mu))
dIdl[2] = 2 / sig ** 2 * (-2 * dIdl[0] - dIdl[1] * (lam - mu))
dIdl[3] = 2 / sig ** 2 * (-3 * dIdl[1] - dIdl[2] * (lam - mu))

# Plot it
fig, ax = plt.subplots(5, figsize=(8, 12))
ax[0].plot(lam, I);
ax[1].plot(lam, dIdl[0]);
ax[2].plot(lam, dIdl[1]);
ax[3].plot(lam, dIdl[2]);
ax[4].plot(lam, dIdl[3]);
ax[0].set_ylabel(r"$I$", fontsize=14)
ax[1].set_ylabel(r"$\frac{dI}{d\lambda}$", fontsize=18);
ax[2].set_ylabel(r"$\frac{d^2I}{d\lambda^2}$", fontsize=18);
ax[3].set_ylabel(r"$\frac{d^3I}{d\lambda^3}$", fontsize=18);
ax[4].set_ylabel(r"$\frac{d^4I}{d\lambda^4}$", fontsize=18);
ax[-1].set_xlabel("Wavelength", fontsize=16);
```

Now, let's compute the derivatives with respect to $\beta$ using the expressions we derived above:

```python
dIdb = np.zeros((4, nwav)) * np.nan

# First deriv
dIdb[0] = -lam * dIdl[0]

# Second deriv
dIdb[1] = lam * dIdl[0] + lam ** 2 * dIdl[1]

# Third deriv
dIdb[2] = -2 * lam * dIdl[0] - 2 * lam ** 2 * dIdl[1] - lam ** 3 * dIdl[2]

# Fourth deriv
dIdb[3] = 6 * lam * dIdl[0] + 8 * lam ** 2 * dIdl[1] + 5 * lam ** 3 * dIdl[2] + lam ** 4 * dIdl[3]
```

Now we compute these derivatives numerically by Doppler-shifting the spectrum by a tiny amount and computing a finite difference:

```python
# Numerical derivatives
lamp = np.zeros(nwav)
lamm = np.zeros(nwav)
beta = 1e-8
for i in range(nwav):
    lamp[i] = (1 + beta) * lam[i]
    lamm[i] = (1 - beta) * lam[i]
dIdb_num = np.empty((4, nwav))

# First deriv
Ip = np.interp(lam, lamp, I)
Im = np.interp(lam, lamm, I)
dIdb_num[0] = (Ip - Im) / (2 * beta)

# Second deriv
Ip = np.interp(lam, lamp, dIdb_num[0])
Im = np.interp(lam, lamm, dIdb_num[0])
dIdb_num[1] = (Ip - Im) / (2 * beta)

# Third deriv
Ip = np.interp(lam, lamp, dIdb_num[1])
Im = np.interp(lam, lamm, dIdb_num[1])
dIdb_num[2] = (Ip - Im) / (2 * beta)

# Fourth deriv
Ip = np.interp(lam, lamp, dIdb_num[2])
Im = np.interp(lam, lamm, dIdb_num[2])
dIdb_num[3] = (Ip - Im) / (2 * beta)
```

Finally, let's compare the expressions:

```python
fig, ax = plt.subplots(5, figsize=(8, 12))
ax[0].plot(lam, I);
ax[1].plot(lam, dIdb_num[0]);
ax[1].plot(lam, dIdb[0], '--');
ax[2].plot(lam, dIdb_num[1]);
ax[2].plot(lam, dIdb[1], '--');
ax[3].plot(lam, dIdb_num[2]);
ax[3].plot(lam, dIdb[2], '--');
ax[4].plot(lam, dIdb_num[3]);
ax[4].plot(lam, dIdb[3], '--');
ax[0].set_ylabel(r"$I$", fontsize=14)
ax[1].set_ylabel(r"$\frac{dI}{d\beta}$", fontsize=18);
ax[2].set_ylabel(r"$\frac{d^2I}{d\beta^2}$", fontsize=18);
ax[3].set_ylabel(r"$\frac{d^3I}{d\beta^3}$", fontsize=18);
ax[4].set_ylabel(r"$\frac{d^4I}{d\beta^4}$", fontsize=18);
ax[-1].set_xlabel("Wavelength", fontsize=18);
```

They agree! (Note that the numerical derivatives become unstable above third order.)
